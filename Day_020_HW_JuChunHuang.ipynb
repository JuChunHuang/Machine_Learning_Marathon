```python
# 載入需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 另一個繪圖-樣式套件
plt.style.use('ggplot')

# 忽略警告訊息
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')

# Notes from example
# enumerate() is a built-in function to iterate through a sequence and keep track of both the index and the number.
# %d will truncate to integer, %s will maintain formatting, %f will print as float and %g is used for generic number
```

```python
# original range: [0, 1); therefore, we subtracted 0.5 then multiplied by 2 --> [-1, 1)
matrix = (np.random.rand(10, 10) - 0.5) * 2

plt.figure(figsize=(10, 10))
heatmap = sns.heatmap(matrix, annot=True)
plt.show()
```

```python
nrow = 1000
ncol = 3

matrix = (np.random.rand(nrow, ncol) - 0.5) * 2

# 隨機給予 0, 1, 2 三種標籤
indice = np.random.choice([0,1,2], size=nrow)
plot_data = pd.DataFrame(matrix, indice).reset_index()

# 繪製 seborn 進階 Heatmap
grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey=False, 
                    hue = 'index', vars = [x for x in list(plot_data.columns) if x != 'index'])

grid.map_upper(plt.scatter, alpha = 0.2)
grid.map_diag(sns.kdeplot)
grid.map_lower(sns.kdeplot, cmap = plt.cm.OrRd_r)

plt.show()
```

```python
nrow = 1000
ncol = 3

matrix = np.random.randn(nrow * ncol).reshape(nrow, ncol)    # randn括號內為產生的隨機數數量

# 隨機給予 0, 1, 2 三種標籤
indice = np.random.choice([0,1,2], size=nrow)
plot_data = pd.DataFrame(matrix, indice).reset_index()

# 繪製 seborn 進階 Heatmap
grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey=False,
                    hue = 'index', vars = [x for x in list(plot_data.columns) if x != 'index'])

grid.map_upper(plt.scatter, alpha = 0.2)
grid.map_diag(sns.kdeplot)
grid.map_lower(sns.kdeplot, cmap = plt.cm.OrRd_r)

plt.show()
```
