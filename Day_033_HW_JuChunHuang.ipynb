### 1. 模型的泛化能力 (generalization) 是指什麼？

    A: 學習到的模型對未知資料的預測能力,通常通過測試誤差來評價學習方法的泛化能力.
### 2. 分類問題與回歸問題分別可用的目標函數有哪些？
    A: 回歸常用的目標函數: Mean square error (MSE) 和 Mean absolute error (MAE)。

    分類問題常用的目標函數: cross-entropy
